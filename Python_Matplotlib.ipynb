# matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

### basic

```python
x = np.linspace(-1, 1, 50)  # use numpy to create x axis
y = 2 * x**2 + 1  # dependent variables
plt.plot(x, y)
plt.show()
```

### figure

```python
x = np.linspace(-3, 3, 50)
y1 = 2 * x + 1
y2 = 4 * x + 3

plt.figure(num=2)
plt.plot(x, y1)

plt.figure(num=3)
plt.plot(x, y2)

plt.show()
```

# matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

### basic

```python
x = np.linspace(-1, 1, 50)  # use numpy to create x axis
y = 2 * x**2 + 1  # dependent variables
plt.plot(x, y)
plt.show()
```

### figure components

```python
x = np.linspace(-3, 3, 50)
y1 = 2 * x + 1
y2 = 4 * x**2 + 3

plt.figure(num=2, figsize=(5,4))  # num: the number of the figure
plt.plot(x, y1)

plt.xlim((-1,4))
plt.ylim((-1,3))

plt.xlabel("x axis")
plt.ylabel("y axis")

plt.yticks([0.5,1,3],['good','medium','bad'])


# the sencond figure
plt.figure(num=3, figsize=(8,6))
plt.plot(x, y2, color='red', linewidth=4, linestyle='--', label='y=4x^2+3')
plt.legend(loc='best')
plt.show()
```

### scatter

```python
n = 1024
t1 = np.random.normal(0, 1, n)
t2 = np.random.normal(0, 1, n)
c = np.arctan2(t2, t1)  # for color value



plt.scatter(t1, t2, s=75, c=c, alpha=0.5)
plt.xlim((-2,2))
plt.ylim((-2,2))

plt.show()
```

### Bar Plot

```python
n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(X, +Y1, facecolor = 'blue')
plt.bar(X, -Y2, facecolor = 'red')

for x,y in zip(X, Y1):
    plt.text(x, y+0.05, '%.2f'%y, ha='center', va='bottom')  # ha: horizontal alignment; va: vertital alignment
for x,y in zip(X, Y2):
    plt.text(x, -y-0.05, '%.2f'%y, ha='center', va='top')  # location of the mark number, x, y, keep 2 digitals

plt.xticks(())  # remove the tickes of x axis
plt.show()
```

# Kit Usage Report Python Try

```python
##  import packages
import pandas as pd
import numpy as np
##  check version
pd.__version__
```

```python
##  read raw data
data0 = pd.read_csv('KitUsageReport_20202021.csv', header = 0)

data0
```

```python
##  copy raw data to data1
data1 = data0.copy()

data1
```

```python
##  change titles
data1.rename(columns = {'Fiscal\nYear':'Year', 
                        'Kit SKU':'SKU', 'Kit Name':'KitName', 
                        'S/N\nCount':'SNcount', 
                        'Shipment\nCount':'ShipmentCount', 
                        'Billing\nCost':'BillingCost',
                        'Billing\nMethod':'BillingMethod',
                        'Rental\nPeriod':'RentalPeriod'}, inplace = True)

##  change index
data1.set_index('SKU', inplace = True)

data1
```

```python
##  change types
print(data1.dtypes)

data1['Year'] = data1['Year'].fillna(0)  #  change all nan to 0
data1 = data1.astype({'Year':int})  #  change year to int

data1
```

```python
#  add a low-turn indicator
data1['LowTurn_20202021'] = np.where(data1['ShipmentCount'] <= 2, 1, 0)
#  move indicator to front
data1.insert(5, 'LowTurn_20202021', data1.pop('LowTurn_20202021'))

data1
```

```python
#  count numbers of low turn
print('The kinds of low turn kits is: ', (data1['LowTurn_20202021'] == 1).sum())
print('The kinds of high turn kits is: ', (data1['LowTurn_20202021'] == 0).sum())
```

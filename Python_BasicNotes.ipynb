# 循环

### while 循环：需要条件

```python
"""
while 条件：
    条件成立代码1
    条件成立代码2
    ...
else：
    循环正常结束以后执行
"""

# 输出0-5的数字。注意格式化输出和运算符

i = 1
while i<4:                   #  有冒号
    print(f'This is {i} and the next one sould be {i+1}.')
    i += 1

```

```python
# 计算100内奇数偶数和

i = 1
odd = 0
even = 0
while i<=100:
    if i % 2 == 0:
        even += i
    else:
        odd += i
    i += 1
print(f'The sum of odd is {odd}\nThe sum of even is {even}')
```

```python
# while 循环使用 else 语句

count = 0
while count < 5:
    print (count, "less than 5")
    count = count + 1
else:
    print (count, "equals 5")
```

### for 循环：知道遍历次数

```python
"""
for 计数器 in 序列 (range(x))：
    条件成立代码1
    条件成立代码2
    ...
else：
    循环*正常结束*以后执行
"""


# 可以直接字符序列

word = 'Cheng Xiaoyue'
print(f'The spelling of {word} is:')
for i in word:
    if i == 'y':
        print('\nI do not like "y".')
        break                             # break 直接全部退出，不运行for的else
    print(i, end='  ')
else:
    print('\nThis is fine')

    
print(f'\nThe spealling of {word} is:')
for i in word:
    if i == 'y':
        print ('\nI do not like "y".')
        continue                          #  continue 只退出当前一次，所以要在之前给计数器加一,运行for的else
    print (i, end = '  ')
else:
    print('\nBut this is fine.')
```

```python
# basic range

for i in range(3,10):
    print(i, end = '\t')
```

### 退出循环 break continue

```python
# continue 只退出当前一次，所以要在之前给计数器加一,运行for的else
# break 直接全部退出，不运行for的else

final = 10
i = 1
while i <= final:
    i += 1
    if (i-1)%3 == 0:
        print(f'This is my love')
        continue
    if i-1 == final:
        print('This is my true love')
        break
    print(f'This is my {i-1} girl friend')
else:
    print('NO girl is my true love')
```

### 三角形

```python
# 三角形

number = 9
j = 1   # 控制第几行
while j<=number:
    i = 1   # 控制第几列
    while i<=j:
        print('*', end='  ')
        i += 1
    print('')
    j += 1
```

### 乘法表

```python
# 乘法表

j = 1   # 后面数字
while j<=9:
    i = 1   # 前面数字
    while i<=j:
        print(f'{i}*{j}={i*j}', end='\t')
        i += 1
    print('')
    j += 1
```

```python
# 切片
'''
序列名[开始位置下标:结束下标位置:步长]
步长默认是1
默认从0开始选取
默认选取到最后
不包含结束下标位置的字符
'''
str0 = '0123456789'
print(str0[2:6:1])      # 2345
print(str0[2:6:2])      # 24
print(str0[:5])         # 01234
print(str0[5:])         # 56789
print(str0[:])          # 0123456789
print(str0[::-1])       # 9876543210
print(str0[-2:-5:-1])   # 876
print(str0[-2:-5], end = 'end')      # 没结果
```

```python
# 字符串查找
# 序列名.find(子串,开始位置下标,结束位置下标）
str1 = 'It is she who damaged my heart is'
print(str1.find('is'))              # 3
print(str1.rfind('is'))             # 31 从右侧开始找
print(str1.index('is', 10, 40))     # 31
print(str1.find('are'))             # -1
print(str1.count('is'))             # 2
```

```python
# 字符串修改
# 序列名.replace(旧子串,新字串,替换次数）
str2 = str1.replace('is', "'s")     # 源字符串数据没有被改变
print(str2)                         # It 's she who damaged my heart 's

# split 分割
str3 = str1.split('is')
str4 = str1.split('is',1)
print(str3, '\n', str4)

# join 合并列表里的字符串
list1 = ['aa', 'bb', 'cc']
list2 = ''.join(list1)
list3 = '..'.join(list1)
print(list2, '\n', list3)
```

```python
# 送出变量

def seqq(n):
    a = 0
    b = 1
    for i in range(n):
        yield a
        a, b = b, a + b

for j in seqq(10):
    print(j, end = ' ')
```

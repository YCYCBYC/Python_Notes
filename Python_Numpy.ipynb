# Numpy Notes

```python
import numpy as np

```

```python
#  priperties of array

array1 = np.array([[1, 2, 3],[2, 3, 4]])
print(array1)
print('number of dim: ', array1.ndim)
print('shape: ', array1.shape)
print('size: ', array1.size)
```

```python
#  define an arry

a = np.array([2, 23, 4])
b = np.array([22, 232, 42], dtype = np.float)
print(a.dtype)
print(b.dtype)
```

```python
#  define all-1 array

c = np.ones((3,4))
print(c)
```

```python
d = np.arange(10, 12, 0.4)
print(d)
```

```python
e = np.linspace(10, 16, 9)
print(e)
```

```python
f = np.arange(2,14).reshape(3,4)
print(f)
```

```python
#  numpy array calculation

print(np.median(f))
print(np.average(f))
```

```python
# show two arrays to indicate the location of none zero items

np.nonzero(f)
```

```python
# sort

np.sort(f)
```
